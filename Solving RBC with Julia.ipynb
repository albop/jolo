# Solving the RBC model with Julia

This worksheet demonstrates how to solve the rbc model using julia. It uses many python libraries, for preprocessing and presentation while actual computations are performed in Julia.

[dolo](http://albop.github.com/dolo/) is used to preprocess the model stored in a .yaml file into julia functions. Julia functions used here (including the notebook) are available on [github](https://github.com/albop/jolo.git)

The [IPython notebook](http://ipython.org) and the [IJulia bridge](https://github.com/JuliaLang/IJulia.jl) are used for this actual notebook.

Inline plots are made using [matplotlib](http://matplotlib.org/) using the [julia wrapper](https://github.com/stevengj/PyPlot.jl).

Note that the collocation algorithm implemented here is written using a fully vectorized approach. As such the timings are probably biased against Julia, since its implementation of vectorized operations is (temporarly) lagging behind.

```julia
model = yaml_import("rbc")
```

# construct approximation space

```julia
# boundaries for the state-space
smin = [1-0.001, 5]
smax = [1+0.001, 10]
orders = [10, 50]

# nodes and weight to compute expectations
nodes = zeros( (1,1) )  # all expected shocks are zero
weights = [1.0]

approx = ApproximationSpace(smin,smax,orders,nodes,weights)

grid = mlinspace(smin,smax,orders)
plot( grid[:,1], grid[:,2], "o" )
xlim( min(grid[:,1]), max(grid[:,1]) )
title("That is a cartesian grid !")
```

```julia
## initial decision rule

# construct cartesian grid
grid = mlinspace(smin,smax,orders)
s_ss = model.calibration["states"]
x_ss = model.calibration["controls"]
X_s = [[ 1.25268335 -0.02185392 ;  0.20023482 -0.00591657]];
N = prod(approx.orders)
init = repmat(x_ss', N, 1) + (grid - repmat(s_ss', N, 1) ) * X_s'
```

```julia
## solve the model using initial decision rule as a starting guess

tic()
dr = solve_model(model, approx, init)
toc()

tic()
dr  = solve_model(model, approx, init)
toc()
```

```julia
# it is also possible to use a decision rule object as an initial guess

# create initial decision rule object
dr = DecisionRule(smin,smax,orders,init);

tic()
dr = solve_model(model, approx, dr)
toc()

tic()
dr = solve_model(model, approx, dr)
toc();
```

# Plot the decision rule


```julia
ssmin = copy(smin)
ssmax = copy(smax)
ssmin[1] = (smin[1] + smax[1])/2
ssmax[1] = (smin[1] + smax[1])/2
svec = mlinspace(ssmin,ssmax,[1,100])   # list of points, where only the second element varies
xvec = evaluate(dr, svec)               # interpolate d.r. on these points;
```

```julia
figure()
for i = 1:size(xvec,2)
    vn = model.symbols["controls"][i]
    plot(svec[:,2], xvec[:,i],label="\$$vn\$")
end
xlabel("\$k_t\$")
legend()
```

# More details about the model object

Here the type of the model is "fga", and is assumed to be in the form

$$ s_t = g(s_{t-1},x_{t-1},y_{t-1},\epsilon_t) $$
$$ f(s_t,x_t,y_t,s_{t+1},x_{t+1},y_{t+1})$$
$$ y_t = a(s_t,x_t) $$

Where $s_t, x_t, y_t$ are called states, controls and auxiliary variables respectively.

The model is coded in the following file.

```julia
run( `cat rbc.yaml` )
```

## functions

All functions of the model follow the same convention. They depend on a many two dimensional arrays, where the first index denotes a point, and on a last 1d vector argument for the parameter values.

```julia
# example to compute auxiliary variables on the grid. Auxiliaries are
# defined by: y = a(s,x,p), where s, x, and p denote the states, the
# controls and the parameters respectively

a = model.functions["auxiliary"]
print(a)

# discretization of the state-space
s = grid 
print(size(s))

# controls on the grid 
x = init
print(size(x))

# parameters values
p = model.calibration["parameters"]
# compute the values
y = a(s,x,p)
```
